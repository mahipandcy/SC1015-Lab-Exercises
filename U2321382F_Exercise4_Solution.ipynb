# Problem 1

```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics


# Import essential models and functions from sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Create a Linear Regression object
linreg = LinearRegression()
```

```python
data = pd.read_csv("train.csv")
print(data.shape)
```

```python
prob1 = data[["GrLivArea", "SalePrice"]]
print("shape: ", data.shape, "\n")
print(prob1.describe())
```

```python
x = data[["GrLivArea"]]
y = data[["SalePrice"]]
```

```python
# (a)Split the Dataset into Train and Test into 80:20 ratio
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
linreg.fit(x_train, y_train)
```

```python
# (a)Regression line

print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)

regline_x = x_train
regline_y = linreg.intercept_ + linreg.coef_ * x_train

f = plt.figure(figsize=(16, 8))
plt.scatter(x_train, y_train)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()
```

```python
# (b)TRAINING SET

y_train_pred = linreg.predict(x_train)

print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_pred))
print()

# (b)TEST SET

y_test_pred = linreg.predict(x_test)

print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(x_test, y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
print("Root Mean Squared Error (MSE) \t:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print()
```

```python
f, axes = plt.subplots(1, 2, figsize=(18, 8))

axes[0].scatter(y_train, y_train_pred, color = "blue")
axes[0].plot(y_train, y_train, 'r-', linewidth = 1)

axes[0].set_xlabel("True values of the Response Variable (Train)")
axes[0].set_ylabel("Predicted values of the Response Variable (Train)")

axes[1].scatter(y_test, y_test_pred, color = "green")
axes[1].plot(y_test, y_test, 'r-', linewidth = 1)

axes[1].set_xlabel("True values of the Response Variable (Test)")
axes[1].set_ylabel("Predicted values of the Response Variable (Test)")

plt.show()
```

Prediction of SalePrice using GrLivArea is the best. For GrLivArea, the R^2 value is closer to one compared to the other two. The difference between the actual test data value and the predicted ones are generally smaller than that of the other two.

For GrLivArea, the MSE value of its prediction on the test set is lower than that of the other two. This means that the predicted value using the regression analysis of GrLivArea deviates lesser from the actual value of the test set data.

Overall, GrLivArea is the best model for predicting sale price.

```python
prob3_data = data[["GrLivArea", "SalePrice"]]
prob3_data.isnull().sum()
```

```python
f = plt.figure(figsize=(16,7))
sb.boxplot(data = data["GrLivArea"] , orient = "h")
f = plt.figure(figsize=(16,7))
sb.boxplot(data = prob3_data["SalePrice"] , orient = "h")
```

```python
Q1 = prob3_data.quantile(0.25)
Q3 = prob3_data.quantile(0.75)
IQR = Q3 - Q1
print("The number of outliers are:")
((prob3_data < (Q1 - 1.5 * IQR)) | (prob3_data > (Q3 + 1.5 * IQR))).sum()
```

```python
for name_var in ['GrLivArea']:
    q75,q25 = np.percentile(prob3_data.loc[:,name_var],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    prob3_data.loc[prob3_data[name_var] < min,name_var] = np.nan
    prob3_data.loc[prob3_data[name_var] > max,name_var] = np.nan
    
for name_var in ['SalePrice']:
    q75,q25 = np.percentile(prob3_data.loc[:,name_var],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    prob3_data.loc[prob3_data[name_var] < min,name_var] = np.nan
    prob3_data.loc[prob3_data[name_var] > max,name_var] = np.nan
    
prob3_data.isnull().sum()
```

```python
prob3_data = prob3_data.dropna(axis = 0)
prob3_data.isnull().sum()
```

```python
print(prob3_data.shape)

grliv = prob3_data[['GrLivArea']]
sale = prob3_data[['SalePrice']]
```

```python
gr_train, gr_test, sale_train, sale_test = train_test_split(grliv, sale, test_size = 0.20)

linreg.fit(gr_train, sale_train)

print("gr_train shape :", gr_train.shape)
print("sale_train shape :", sale_train.shape)
print("gr_test shape :", gr_test.shape)
print("sale_test shape :", sale_test.shape)
```

```python
print('Intercept of Regression \t: b = ', linreg.intercept_)
print('Coefficients of Regression \t: a = ', linreg.coef_)

regline_gr = gr_train
regline_sale = linreg.intercept_ + linreg.coef_ * gr_train

f = plt.figure(figsize=(16, 8))
plt.scatter(gr_train, sale_train)
plt.plot(regline_gr, regline_sale, 'r-', linewidth = 3)
plt.show()
```

```python
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(gr_train, sale_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(regline_gr, regline_sale))
print()
```

```python
sale_test_pred = linreg.predict(gr_test)

regline_gr_pred = gr_test
regline_sale_pred = linreg.intercept_ + linreg.coef_ * gr_test

f = plt.figure(figsize=(16, 8))
plt.scatter(gr_test, sale_test)
plt.plot(regline_gr_pred, regline_sale_pred, 'r-', linewidth = 3)
plt.show()
```

```python
print("Goodness of Fit of Model \tTest Dataset")
print("Explained Variance (R^2) \t:", linreg.score(gr_test, sale_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(regline_gr_pred, regline_sale_pred))
print()
```
