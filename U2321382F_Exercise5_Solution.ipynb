```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
sb.set() 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import plot_tree
```

# Problem 1

```python
data = pd.read_csv('train.csv')
```

```python
ctrair = pd.DataFrame(data['CentralAir'])
saleprice = pd.DataFrame(data['SalePrice'])

X_train, X_test, y_train, y_test = train_test_split(saleprice, ctrair, test_size = 0.25)

joint = pd.concat([X_train, y_train], axis = 1).reindex(X_train.index)

f = plt.figure(figsize=(18, 6))
sb.boxplot(x = "SalePrice", y = "CentralAir", data = joint, orient = "h")
```

```python
#Tree of Depth 3
dectree = DecisionTreeClassifier(max_depth = 3)  
dectree.fit(X_train, y_train)                    

y_train_pred = dectree.predict(X_train)
y_test_pred = dectree.predict(X_test)

tn_train, fp_train, fn_train, tp_train = confusion_matrix(y_train, y_train_pred, labels=['N', 'Y']).ravel()
print(tn_train, fp_train, fn_train, tp_train)

tn_test, fp_test, fn_test, tp_test = confusion_matrix(y_test, y_test_pred, labels=['N', 'Y']).ravel()
print(tn_test, fp_test, fn_test, tp_test)


print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", dectree.score(X_train, y_train))
print("True postive rate \t\t:", (tp_train/(tp_train+fn_train)))
print("False positive rate \t\t:", (fp_train/(tn_train+fp_train)))
print()


print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", dectree.score(X_test, y_test))
print("True postive rate \t\t:", (tp_test/(tp_test+fn_test)))
print("False positive rate \t\t:", (fp_test/(tn_test+fp_test)))
print()
print()
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y_train, y_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])

f = plt.figure(figsize=(12,12))
plot_tree(dectree, filled=True, rounded=True, 
          feature_names=["SalePrice"], 
          class_names=["CentralAir:No","CentralAir:Yes"])
plt.show()
```

```python
#Tree of Depth 4
dectree = DecisionTreeClassifier(max_depth = 4)  
dectree.fit(X_train, y_train)                    

# Predict CentralAir
y_train_pred = dectree.predict(X_train)
y_test_pred = dectree.predict(X_test)

tn_train, fp_train, fn_train, tp_train = confusion_matrix(y_train, y_train_pred, labels=['N', 'Y']).ravel()
print(tn_train, fp_train, fn_train, tp_train)

tn_test, fp_test, fn_test, tp_test = confusion_matrix(y_test, y_test_pred, labels=['N', 'Y']).ravel()
print(tn_test, fp_test, fn_test, tp_test)


print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", dectree.score(X_train, y_train))
print("True postive rate \t\t:", (tp_train/(tp_train+fn_train)))
print("False positive rate \t\t:", (fp_train/(tn_train+fp_train)))
print()


print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", dectree.score(X_test, y_test))
print("True postive rate \t\t:", (tp_test/(tp_test+fn_test)))
print("False positive rate \t\t:", (fp_test/(tn_test+fp_test)))
print()
print()
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y_train, y_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])

f = plt.figure(figsize=(12,12))
plot_tree(dectree, filled=True, rounded=True, 
          feature_names=["SalePrice"], 
          class_names=["CentralAir:No","CentralAir:Yes"])
plt.show()
```

The model with depth 4 demonstrates superior performance. On the train dataset, the depth 4 model outshines the depth 3 model with a higher classification accuracy (0.946 vs. 0.944), a slightly improved true positive rate (0.988 vs. 0.990), and a notably lower false positive rate (0.662 vs. 0.718). This trend continues on the test dataset, where the depth 4 model maintains its advantage with a higher classification accuracy (0.918 vs. 0.934), a superior true positive rate (0.968 vs. 0.982), and a lower false positive rate (0.792 vs. 0.750).

Although YearBuilt exhibits the lowest false positive rate on the train dataset (0.704), SalePrice's false positive rate (0.744) remains competitive. However, on the test dataset, OverallQual surpasses with a lower false positive rate (0.823) compared to SalePrice (0.882) and YearBuilt (0.833). Despite these nuances, SalePrice emerges as the superior predictor due to its consistently high classification accuracy and commendable true positive rate across both datasets, making it the optimal variable for prediction purposes.
